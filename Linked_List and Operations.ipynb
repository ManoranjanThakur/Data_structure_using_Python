```python
class Node:
    def __init__(self,data):
        self.data=data
        self.ref=None
```

```python
class LinkedList:
    def __init__(self):
        self.head=None
        
#--------Traversing the linked List-------

    def print_LL(self):
        if self.head is None:
            print("Linked List is empty")
        else:
            n=self.head
            while n is not None:
                print(n.data)
                n=n.ref
                
                
                
                
#--------Adding elements in the Linked List--------
    
    #1-> Adding at the start
    def add_start(self,data):
        new_node = Node(data)
        new_node.ref= self.head
        self.head=new_node
    
    #2-> Adding at the end
    def add_end(self,data):
        new_node=Node(data)
        if self.head is None:
            self.head=new_node
        else:
            n=self.head
            while n.ref is not None:
                n=n.ref
            n.ref=new_node
            
    #3-> Adding in between
    # While adding element in between node we have two condition
    #  whether inserting the node after any specific node or before
    # any specific node
    
    #3-a> After a given node
    def after_node(self,data,x):
        n=self.head
        while n is not None:
            if(n.data==x):
                break
            n=n.ref
        if(n is None):
            print("Element not found")
        else:
            new_node=Node(data)
            new_node.ref=n.ref
            n.ref=new_node
            
    #3-b> Before a given node
    def before_node(self,data,x):
        if self.head is None:
            print("Linked List is Empty")
            return
        elif self.head.data==x:
            new_node=Node(data)
            new_node.ref=self.head
            self.head=new_node
        else:
            n=self.head
            while n.ref is not None:
                if(n.ref.data==x):
                    break
                n=n.ref
            if n.ref is None:
                print("Element not Found")
                return
            new_node=Node(data)
            new_node.ref=n.ref
            n.ref=new_node  
            
            
#-----------Deleting elements from the Linked List----------

    #-1-> Deleting element from the beginning of the LinkedList
    def delete_start(self):
        if self.head is None:
            print("List is already empty")
        else:
            self.head=self.head.ref
            
    #-2-> Deleting element from the end of the LinkedList
    def delete_end(self):
        n=self.head
        if n is None:
            print("Empty List")
        elif n.ref is None:
            n=None
        else:
            while(n.ref.ref is not None):
                n=n.ref
            n.ref = None
            
    #-3-> Deleting element from any specific position in the LinkedList
    def delete_value(self,x):
        self.head
        if(self.head is None):
            print("Linked List is empty")
        elif (self.head.data==x):
            n=None
        else:
            while(self.head.ref is not None):
                if(self.head.ref.data==x):
                    break
                n=n.ref
            if self.head.ref is None:
                print("Value not present")
                return
            self.head.ref=self.head.ref.ref
```

```python
ll1=LinkedList()
ll1.add_start(10)
ll1.add_start(20)
ll1.add_start(30)
ll1.add_end(100)
ll1.after_node(110,20)
ll1.before_node(130,110)
ll1.print_LL()
ll1.delete_end()
ll1.print_LL()
```
